# **Indrodução**

O Índice de Desenvolvimento da Educação Básica (Ideb) foi criado em 2007 e reúne os resultados de dois conceitos importantes para a qualidade da educação: o fluxo escolar e as médias de desempenho nas avaliações.

## Objetivo:

O Ideb tem como objetivo medir o desempenho dos estudantes em duas etapas da educação básica: o Ensino Fundamental (anos iniciais e finais) e o Ensino Médio.

## Avaliações: 

As avaliações utilizadas para o cálculo do Ideb são a Prova Brasil, aplicada para alunos do 5º e 9º ano do Ensino Fundamental, e o Exame Nacional do Ensino Médio (ENEM), aplicado para estudantes do 3º ano do Ensino Médio e a partir dos resultados dessas avaliações e da taxa de aprovação, é possível calcular o índice para cada escola, município, estado e para o país como um todo.

O Ideb é expresso em uma escala de 0 a 10, sendo que o valor 6 é considerado a meta nacional estabelecida para o Brasil até o ano de 2022.

## Políticas Públicas para a Educação:

Além de permitir a avaliação do desempenho educacional, o Ideb também serve como base para a formulação de políticas públicas e para o estabelecimento de metas de qualidade na educação, alem de ter sido uma importante ferramenta para acompanhar a evolução da qualidade da educação no país, identificar desafios e promover melhorias.
Ele ajuda a mobilizar gestores, educadores, pais e alunos em prol de uma educação básica de qualidade e igualdade de oportunidades para todos os estudantes brasileiros.

## Explicando a tabela:
A tabela pussui dados dos anos de 2005 á 2021, sobre as redes publicas, estaduais e privada com os ensinos fundamental e medio, nela conseguimos analisar a taxa de aprovação, indicador de rendimento, notas em materias de português e matemática da saeb (Sistema Nacional de Avaliação da Educação Básica) saeb é um conjunto de avaliações externas em larga escala que permite ao Inep realizar um diagnóstico da educação básica brasileira e de fatores que podem interferir no desempenho do estudante.

## No programa a seguir retorna os dados na tela através de uma tabela

### Aqui o usuário terá 3 alternativas, são elas:
    
O usuário escolhe como ele deseja pesquisar a tabela escolhendo a opção 1:rede, que permite pesquisar entre as redes publicas, estaduais e privadas ou escolher a opção 2:ensino, que irá buscar os dados do ensino fundamental ou medio e uma terceira opção que é sair e não pesquisar nenhuma tabela.

### Salvar a tabela:

Em seguida irá ser perguntado se o usuário deseja salvar a tabela "Deseja salvar a tabela? (s/n)" s(sim) e n (não), com o sim ele irá escolher o diretório de destino e salvar a tabela aonde desejar.

Logo a pós salvar ele poderar escolher entre pesquisa novamente ou finalizar a consulta.

```python
import pandas as pd
from IPython.display import display
import tkinter as tk
from tkinter import filedialog

df = pd.read_csv('projeto.edu.av2.csv', encoding='latin1')

while True:
    # Solicitar ao usuário a pesquisa desejada
    print("Escolha a opção de pesquisa:")
    print("1. Pesquisar por rede")
    print("2. Pesquisar por tipo de ensino")
    print("0. Sair")
    
    opcao = input("Opção: ")
    
    if opcao == "1":
        rede_desejada = input("Digite a rede desejada (privada, estadual ou publica): ")
        df_filtered = df.loc[df['rede'] == rede_desejada]
        coluna_pesquisada = 'rede'
        valor_pesquisado = rede_desejada
    elif opcao == "2":
        ensino_desejado = input("Digite o tipo de ensino desejado (fundamental ou medio): ")
        df_filtered = df.loc[df['ensino'] == ensino_desejado]
        coluna_pesquisada = 'tipo de ensino'
        valor_pesquisado = ensino_desejado
    elif opcao == "0":
        break
    else:
        print("Opção inválida. Tente novamente.")
        continue

    if df_filtered.empty:
        print(f"Não há dados disponíveis para {coluna_pesquisada} '{valor_pesquisado}'.")
    else:
        display(df_filtered)

        # Programa para salvar a tabela onde o usuário desejar:
        resposta = input("Deseja salvar a tabela? (s/n): ")

        if resposta.lower() == 's':
            root = tk.Tk()
            root.withdraw()
            folder_selected = filedialog.askdirectory()

            if folder_selected:
                file_path = f"{folder_selected}/tabela_{coluna_pesquisada}_{valor_pesquisado}.csv"
                df_filtered.to_csv(file_path, index=False, encoding='utf-8')
                print(f"Tabela salva em: {file_path}")
            else:
                print("Nenhum diretório selecionado. A tabela não foi salva.")
        else:
            print("A tabela não foi salva.")

        print()

print("Programa encerrado.")
```

## Gráfico da taxa de aprovação

Aqui podemos ver um gráfico comparando a taxa de aprovação do ensino fundamental e medio das escolas estaduais, privadas e publicas.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('projeto.edu.av2.csv', encoding='latin1')

# Filtrar os dados por uma determinada pesquisa, por exemplo, rede
rede_desejada = input("Digite a rede desejada (privada, estadual ou publica): ")
df_filtered = df.loc[df['rede'] == rede_desejada]

if df_filtered.empty:
    print(f"Não há dados disponíveis para a rede {rede_desejada}.")
else:
    # Criar o gráfico de barras da taxa de aprovação
    plt.bar(df_filtered['ensino'], df_filtered['taxa_aprovacao'])
    plt.xlabel('Escola')
    plt.ylabel('Taxa de Aprovação')
    plt.title(f'Taxa de Aprovação por Escola ({rede_desejada})')
    plt.xticks(rotation=90)
    plt.show()
```
